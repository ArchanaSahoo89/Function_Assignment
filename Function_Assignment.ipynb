```python
# 2. Explain the concept of function arguments and parameters in Python.

# In Python, parameters and arguments are terms related to functions but serve different purposes:

# I) Parameters:

# These are the variables listed inside the parentheses in the function definition.
# Parameters act as placeholders for the values that are passed into the function when it's called.
# They define the inputs that the function can receive.
# Parameters are variables in the function definition.

#Example:

def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")


# II) Arguments:

# These are the actual values passed to the function when it is called.
# Arguments are assigned to the corresponding parameters in the function definition.
# Arguments are the actual values passed to the function.

#Example:

greet("Arman") # 'name' is a arguments

when we call greet("Arman"), the argument "Arman" is passed to the parameter name.


# Types of Arguments and Parameters in Python:

# Python allows different ways to pass arguments and define parameters in functions:

# i) Positional Arguments:

# These are arguments passed to a function in the correct position order.
# The first argument is assigned to the first parameter, the second to the second, and so on.

#ex:
def addn(a, b):
    return a + b

result = addn(3, 5)  # 3 and 5 are positional arguments
print(result)

# ii) Keyword Arguments:

# These are arguments passed by explicitly naming the parameter in the function call.
# The order of arguments doesn't matter when using keyword arguments.

#ex:
result = add(b=5, a=3)  # Using keyword arguments
print(result)

# iii) Default Parameters:

# we can define default values for parameters in a function.
# If no argument is provided for a parameter with a default value, the function will use the default.

#ex:
def greet(name="World"):
    print(f"Hello, {name}!")

greet()
greet("Arman")

# iv) Arbitrary Arguments (*args):

# Allows to pass a variable number of positional arguments.
# The *args parameter collects extra arguments into a tuple.

#ex:
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))


# v) Arbitrary Keyword Arguments (**kwargs):

# Allows passing a variable number of keyword arguments.
# The **kwargs parameter collects extra keyword arguments into a dictionary.

#ex:
def describe_person(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

describe_person(name="Alice", age=25, city="New York")
```

```python
# 4. What is the purpose of the `return` statement in a Python function?

# The return statement in a Python function is used to:

# Exit the Function: The return statement immediately terminates the function's execution and exits it.

# Return a Value: It allows the function to pass data back to the caller. This data can be of any type (integer, string, list, etc.) or even a complex #object.

# No Return Value (Optional): If no return statement is used, or if return is used without a value, the function implicitly returns None.

# Key Purposes of the return Statement:

# i) Returning a Single Value

# The most common use is returning a computed value to the caller.

#ex:

def square(x):
    return x * x

result = square(5)
print(result)

# ii) Returning Multiple Values

# In Python, a function can return multiple values as a tuple.
#ex:

def get_coordinates():
    return 10, 20

x, y = get_coordinates()
print(x, y)

# iii) Early Exit from a Function

# The return statement can be used to exit a function early, even before reaching the end of the function body.

#ex:

def check_positive(n):
    if n < 0:
        return "Negative number"
    return "Positive number"

print(check_positive(5))
print(check_positive(-3))

# iv) Returning Nothing (None)

# If a function has no return statement or if return is used without a value, Python will return None by default.

#ex:

def greet():
    print("Hello")

result = greet()
print(result)

# v) Returning a Function (Higher-Order Functions)

# In advanced cases, functions can return other functions.

#ex:
def outer_function():
    def inner_function():
        return "Hello from inner function"
    return inner_function

result_function = outer_function()
print(result_function())


# The return statement is essential for passing data from a function to the user.
```

```python
# 5. What are iterators in Python and how do they differ from iterables?

# In Python, iterators and iterables are closely related concepts, but they are not the same.

# I) Iterable

# An iterable is any Python object that can return its elements one at a time, allowing it to be looped over. Examples of iterables include lists, tuples, strings, dictionaries, sets, and more.
# An object is considered iterable if it implements the __iter__() method, which returns an iterator.
# You can iterate over an iterable using a for loop or by explicitly getting an iterator using the iter() function.
# Iterable: An object that can be iterated over (like lists, tuples, etc.) and produces an iterator using iter().

#ex:
my_list = [1, 2, 3]
for item in my_list:  # Iterating over the iterable
    print(item)

# II) Iterator

# An iterator is an object that represents a stream of data. It produces the items from an iterable, one at a time, on demand.

# An object is an iterator if it implements both the __iter__() and __next__() methods.
# The __iter__() method returns the iterator object itself.
# The __next__() method returns the next item from the iterator. If there are no more items, it raises a StopIteration exception to signal the end.
# Once an iterator is exhausted (i.e., there are no more items), it cannot be reset or reused without recreating it.
# Iterator: An object that produces one item at a time using next() until it is exhausted. An iterator can only be used once unless it is recreated.

#Example of an Iterator:

my_list = [1, 2, 3]
my_iterator = iter(my_list)


# All iterators are iterables, but not all iterables are iterators.
```

```python
# 6. Explain the concept of generators in Python and how they are defined.

# In Python, generators are a special type of iterator that allow you to iterate over a sequence of values lazily (one at a time) without storing the entire sequence in memory at once.
# Generators are highly efficient for large data sets or infinite sequences because they generate values on demand rather than all at once.
# Generators are similar to regular functions but use the yield keyword instead of return to produce values one at a time.
# When a generator function is called, it returns a generator object, which can be iterated over using a loop or the next() function.

# Characteristics

# Lazy Evaluation: Generators generate values one at a time when requested, rather than computing and storing the entire sequence in memory.
# State Retention: After each yield, the state of the generator (local variables, execution point) is saved, and it resumes from that state when the next value is requested.
# Exhaustion: Generators can only be iterated once. Once all values are generated, they cannot be reset unless the generator function is called again to create a new generator object.
# Automatic StopIteration: When there are no more values to yield, the generator automatically raises a StopIteration exception.


# Defining a Generator Function:

# A generator is defined like a normal function but uses yield to return values lazily.
#Each time the generator's __next__() method is called, the function runs until it hits a yield statement, and then it pauses its state, returning the value. When __next__() is called again, the function resumes execution right after the yield.

#ex:
def count_up_to(max_value):
    num = 1
    while num <= max_value:
        yield num  # Yield the current value of num
        num += 1
# cretaing a generator object

counter = count_up_to(3)

# Iterating over the generator object
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
# The next call would raise StopIteration as there are no more values

# Generator expressions:

# Generator expressions provide a concise way to define generators.

# Python supports a shorthand for defining simple generators, known as generator expressions. These are similar to list comprehensions but use parentheses instead of square brackets.

#ex:

gen_exp = (x * x for x in range(4))

# Using a generator expression
for value in gen_exp:
    print(value)  # Output: 0, 1, 4, 9


# Advantages of Generators

# Memory Efficiency: Generators are memory-efficient because they generate values on the fly rather than storing the entire sequence in memory.

# Infinite Sequences: Generators are perfect for representing infinite sequences since they don't compute the entire sequence upfront.

#Example: Generating an infinite sequence of Fibonacci numbers.

def infinite_fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Improved Performance: For large data sets or expensive computations, generators avoid the overhead of building entire lists or other structures in memory.
```

```python
# 8. What is a lambda function in Python and when is it typically used?

# A lambda function in Python is a small, anonymous function defined using the lambda keyword.
# Unlike regular functions defined using def, lambda functions can have any number of arguments but only a single expression. They are typically used when you need a short, throwaway function for a brief task.

# Characteristics of Lambda Functions

# Anonymous: Lambda functions are not given a name by default, although they can be assigned to a variable.
# Single Expression: A lambda function contains a single expression, which is evaluated and returned. It cannot contain multiple statements or complex logic.
# Inline Use: Lambda functions are typically used inline where short, simple functions are needed temporarily.

# The syntax for a lambda function is:

# lambda arguments: expression

#ex:
add = lambda x, y: x + y

print(add(3, 5))

# Lambda: The keyword that defines the lambda function.
# Arguments: A comma-separated list of arguments (can be zero or more).
# Expression: A single expression that is evaluated and returned (implicitly).

# Lambda functions are often used in places where a small function is required for a short period and defining a full function using def would be overkill.

# I) Sorting and Filtering

# Lambda functions are frequently used as key functions for operations like sorting or filtering.

#Example: Sorting with sorted()

points = [(1, 2), (4, 1), (5, -1)]
# Sort by the second element of each tuple

sorted_points = sorted(points, key=lambda x: x[1])
print(sorted_points)

# Example: Filtering with filter()

numbers = [1, 2, 3, 4, 5, 6]

# Filter out even numbers

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4, 6]

# II) Used in Higher-Order Functions

# Lambda functions are often passed as arguments to higher-order functions (functions that take other functions as arguments). Common examples include map(), filter(), and reduce().

#ex: [Using reduce()]

from functools import reduce

numbers = [1, 2, 3, 4]

# Calculate the product of all numbers

product = reduce(lambda x, y: x * y, numbers)
print(product)

# III)  Creating Small Throwaway Functions

# Lambda functions are useful when you need a small, quick function for a one-time task and don't want to define a full function using def.

#Example: Passing a lambda directly to double a number

print((lambda x: x * 2)(5))  # Output: 10

# IV) Default Sorting

# Lambda functions are often used to provide a custom sorting logic without the need for defining a separate function.

#Example: Custom sorting

words = ['apple', 'banana', 'cherry']

# Sort by the length of the word

words_sorted = sorted(words, key=lambda word: len(word))
print(words_sorted)

# Lambda fuctions are Restricted to single expressions, less readable for complex logic, and can’t include statements like if, for, or while. For more complex logic, regular functions should be used.
```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

# The map() function in Python is used to apply a given function to each item of an iterable (such as a list, tuple, or string) and return an iterator (or a list in Python 2) that contains the results of applying the function to each element.

# Syntax of map():

# map(function, iterable[, iterable2, iterable3, ...])

# function: The function that will be applied to the items of the iterable. This can be a built-in function, a user-defined function, or a lambda function.
# iterable: The sequence (e.g., list, tuple) whose items the function will be applied to. You can pass multiple iterables as well, and the function will be applied to corresponding elements of the iterables.

#Ex:

# A simple function to square a number
def square(num):
    return num ** 2

# Applying the function to a list of numbers
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

# Convert the map object to a list to see the results
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]


# How map() Works:

# map() applies the given function to each item in the iterable.
# It returns an iterator (which you can convert to a list or tuple, for example).
# If multiple iterables are passed, the function must take that many arguments, and map() will apply the function to corresponding items from each iterable.

#Examples of map() Usage

# I) Single Iterable

# Here’s how map() works with a single iterable:

# A simple function to square a number
def square(num):
    return num ** 2

# Applying the function to a list of numbers
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

# Convert the map object to a list to see the results
print(list(squared_numbers))

#II) Using lambda with map()

# we can also use a lambda function directly in map() for short, anonymous functions.

numbers = [1, 2, 3, 4, 5]

# Using lambda to cube each number
cubed_numbers = map(lambda x: x ** 3, numbers)

print(list(cubed_numbers))

# III) Multiple Iterables

# If you pass multiple iterables to map(), the function you provide must accept that many arguments. map() will apply the function to the corresponding elements from each iterable.

# A function to add two numbers
def add(x, y):
    return x + y

# Two lists to apply the function to
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Apply the add function to corresponding elements of both lists

result = map(add, list1, list2)
print(list(result))

# 4. Converting a map object to a list

# Since map() returns an iterator in Python 3, we convert it into a list or tuple to see the results immediately.

# Squaring numbers in a list
numbers = [1, 2, 3]
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert to a list to see the output
print(list(squared_numbers))

numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))


# Common Use Cases for map()

# Transforming Data: You can easily transform data by applying a function to each element of an iterable. For example, converting a list of strings to uppercase:

words = ['hello', 'world']
upper_words = map(lambda word: word.upper(), words)
print(list(upper_words))  # Output: ['HELLO', 'WORLD']


# Performing Operations on Multiple Lists: When you need to combine or compare elements from multiple lists, map() provides a clear and concise solution.
# Multiply corresponding elements of two lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = map(lambda x, y: x * y, list1, list2)
print(list(result))

# Replacing Loops for Efficiency: When processing large amounts of data, using map() can be more efficient than using traditional for loops, especially when combined with other functional programming tools like filter() or reduce().
```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

# In Python, map(), reduce(), and filter() are functional programming tools used for transforming or processing collections of data like lists, tuples, etc. Although they are all used to work with iterables, each function has a distinct purpose and behavior.

# I) map() Function

# Purpose: Applies a given function to each item of an iterable (e.g., list, tuple) and returns an iterator containing the results.
# Usage: Used when you need to transform all items in an iterable.

# Example:1

numbers = [1, 2, 3, 4]

# Using map to square each number in the list
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16]

# In this example, the lambda function x ** 2 is applied to each element of numbers, and the result is an iterator with squared numbers.

#Example:2

# If multiple iterables are passed, map() applies the function to corresponding elements of each iterable.

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Using map to add corresponding elements of two lists

summed_numbers = map(lambda x, y: x + y, list1, list2)
print(list(summed_numbers))


# II) reduce() Function
# Purpose: Applies a given function to pairwise elements of an iterable, cumulatively reducing the iterable to a single value.

# Usage: Used when you want to aggregate or reduce an iterable to a single result. For example, summing a list of numbers or finding a product of numbers.

# reduce() is part of the functools module and needs to be imported.

#Example:

from functools import reduce

numbers = [1, 2, 3, 4]

# Using reduce to find the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
# Here, reduce() cumulatively multiplies the elements of numbers, resulting in 24.


# III) filter() Function

#Purpose: Applies a filtering function to an iterable, keeping only the elements for which the function returns True.
#Usage: Used to filter out unwanted elements from an iterable based on a condition.
#Example:

numbers = [1, 2, 3, 4, 5, 6]

# Using filter to keep only even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]

# In this example, the lambda function checks whether each element is even, and filter() returns only the even numbers.


# Comparison of Usage:
# map():

# Transformation: Use map() when you need to apply a transformation to all elements in an iterable.
# Example: Squaring numbers, converting strings to uppercase, etc.

# reduce():

# Aggregation: Use reduce() when you want to reduce an iterable to a single value by applying a cumulative function.
# Example: Summing or multiplying all numbers, finding the maximum or minimum, etc.

# filter():

# Filtering: Use filter() when you want to select elements that satisfy a given condition from an iterable.
# Example: Filtering out odd numbers, keeping only positive values, etc.


# Each of these functions plays a specific role in processing data, and they can be very powerful when used appropriately in different situations.
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(num):

    if num < 2:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")
    else:
        print(f"{number} is not a prime number")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.i = 0  # Current index
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        if self.i == 0:
            self.i += 1
            return self.a
        elif self.i == 1:
            self.i += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.i += 1
            return self.b

fib = Fibonacci(10)
for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

list_of_tuples = [(1, 3), (4, 1), (2, 2), (5, 0)]

sorted_list = sorted(list_of_tuples, key=lambda x: x[1])

print(sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):

    is_not_vowel = lambda char: char.lower() not in 'aeiou'

    filtered_string = ''.join(filter(is_not_vowel, input_string))

    return filtered_string


example_string = "Hello, World!"
print(remove_vowels(example_string))
```

```python
# 11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

# Write a Python program using lambda and map.

orders = [
    (34587, 4, 40.95),
    (98762, 5, 56.80),
    (77226, 3, 32.95),
    (88112, 2, 24.99)
]

calculate_order_value = lambda order: (order[0], order[1] * order[2] + 10 if order[1] * order[2] < 100 else order[1] * order[2])

adjusted_orders = list(map(calculate_order_value, orders))

print(adjusted_orders)
```
